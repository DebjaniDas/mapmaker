```python
import pandas as pd
import numpy as np

%matplotlib notebook 
import matplotlib.pyplot as plt

# %matplotlib inline 
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
df = pd.read_csv('planets.csv')
```

```python
exodf = df.loc[(df['st_mass']<=0.6) & (df['st_teff']<=6000)]
```

```python
ra = exodf['ra'] #Right Ascenscion, lattitudinal coordinate
dec = exodf['dec'] #Declination, longitudinal coordinate
distance = exodf['st_dist'] #distance to star, radial coordinate
```

```python
x = distance * np.cos(dec) * np.cos(ra)
y = distance * np.cos(dec) * np.sin(ra)
z = distance * np.sin(dec)
```

```python
mass = exodf['st_mass']
teff = exodf['st_teff']
```

```python
#Add "%matplotlib notebook" before all matplotlib imports in the top cell

fig = plt.figure(figsize=[10, 8]) 
# ^ This controls the size of the figure

ax = fig.add_subplot(111, facecolor='Black', projection='3d')
# ^ This gives the background color

ax.scatter(x, y, z, c=teff, cmap='plasma', s=5*((10*mass)**2), marker='*')   # This plots all the points as stars


# ax.set_xlim(-250,250)
# ax.set_ylim(-250,250)
# ax.set_zlim(-250,250)

ax.set_axis_off()
```
